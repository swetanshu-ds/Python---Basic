###  create the model

```python
#  sequential model is a feed forward model  i.e,  values are always going from left to right
```

```python
model  = tf.keras.models.Sequential()
```

```python
#   but sequential  model are 28*28 so we need to convert it into 1*784
```

```python
##   input into 1 * 784
model.add(tf.keras.layers.Flatten())
```

```python
#  adding few layers
model.add(tf.keras.layers.Dense(128,activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(128,activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(128,activation = tf.nn.relu))

##  dense connect all the cells to the next layer  or fully connected cells
```

```python
#  Output layer
model.add(tf.keras.layers.Dense(10,activation = tf.nn.softmax))

#  we wrote 10 above as we want only possibilities from  0 to 9  and softmax is probability  function  
#  softmax is a probility function will give who so ever the probability of function is more it will give the the number 
```

```python
##  methods 
### step I---
# 1.  import the library as tensorflow
# 2.  import the data set  using mnist 
# 3.  divide  the dataset into training and test data set
# 4.  training model on the training set
# 5.  predicting the Test set result
# 6.  visualizing the test set result
```

```python
model.compile(optimizer = "adam",loss = "sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,epochs = 3)   ##  epochs is the number of time  we want to print
```

```python
val_loss , val_acc = model.evaluate(x_test,y_test)
```

```python
#  let us see we can make good observation or not
predictions = model.predict(x_test)
```
