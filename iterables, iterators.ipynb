```python
c = [1,23,5,7,8,9]   #list type
for i in c:
    print(i)
```

```python
d = (1,2,3,5)   #type tuple
print(type(d))
for i in d:
    print(i)
```

```python
e = 12.5 #float type
for i in e:
    print(i)
```

```python
f = {1,2,3,4,4,4,4,5}
print(type(f),f)
for i in f:
    print(i)
```

```python
g = {"name": "John",
    "age": 23,
    "gender":"Male"}
for item in g:
    print(item)
for item in g.values():
    print(item)
for item in g.keys():
    print(item)
for item in g.items():
    print(item)
```
