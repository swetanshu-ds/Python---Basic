###  generators

```python
###  generators are a kind of functions which can 
###  send back a value and can later resume to  pick 
### it up where it left up    
```

```python
def cubes(n):
    for i in range(n):
         yield i**3
```

```python
for x in cubes(5):
        print(x)
```

```python
class Circle:
    pi = 3.14
    
    def _init_(self,radius=1):
        self.radius = radius
        self.area =  radius*radius*Circle.pi
    def  setradius(self,new_radius):
         self.radius = new_radius
         self.area = new_radius*new_radius*self.pi
    def  circumference(self):
        return  self.radius*self.pi*2
c=Circle()
print("AREA IS ",c.area)
```
