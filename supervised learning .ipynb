##  prediction using supervised learning

```python
#  to predict the percentage of marks of an student based on the number of study hours
```

```python
# importing all librairies required in this notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#  reading the data from gven link
url = "http://bit.ly/w-data"
s_data = pd.read_csv(url)
print("data imported successfully ")
s_data.head(10)                                    #  10 is number of rows 
```

```python
##   lets plot our data points on 2-D graph to eyeball  our dataset and 
##  see if we can manually find any relationship between te data.we can 
##   create the plot with the following script
```

```python
##   plotting  the distribution of scores
s_data.plot(x = "Hours", y = "Scores",style = "o")
plt.title("Hours vs Percentage")
plt.xlabel("Hours Studied ")
plt.ylabel("Percentage Score")
plt.show()
```

```python
##   plotting  the distribution of scores
s_data.plot(x  = "Hours", y = "Scores",style = "--")
plt.title("Hours vs Percentage")
plt.xlabel("Hours Studied ")
plt.ylabel("Percentage Score")
plt.show()
```

```python
##  from the graph above , we can clearly see that there is a positive
#   linear relation between the number of hours studied and percentage of scores
```

### preparing  the data

```python
#  the next step is to divide the data into "attributes"(inputs)  and "labels"(outputs)
```

```python
X = s_data.iloc[:, :-1].values
y = s_data.iloc[:, 1].values
```

```python
y,X
```

```python
#  data preparation
x = s_data[['Hours']].values
y = s_data[['Scores']].values
```

```python
x,y
```

```python
#  now that we have  our attributes and labels   the next 
# step is to split  this data into training and test sets.
#  We'll do this by using Scikit-Learn built-in  train_test_split()  method:
```

```python
"from sklearn.model_selection import train_test_split  \n",
"X_train, X_test, y_train, y_test = train_test_split(X, y, \n",
"                            test_size=0.2, random_state=0) "
```

###  training the Algorithm

```python
###  we have split our data into training and testing sets ,
###  and now is finally time to train our algorithm.
```

```python
"from sklearn.linear_model import LinearRegression  \n",
"regressor = LinearRegression()  \n",
"regressor.fit(X_train, y_train) \n",
"\n",
"print(\"Training complete.\")"
```

```python
###  plotting the regession line

"line = regressor.coef_*X+regressor.intercept_\n",
"\n",
"# Plotting for the test data\n",
"plt.scatter(X, y)\n",
"plt.plot(X, line);\n",
"plt.title(\"Comparison\")\n",
"plt.show()"
```

###  making predictions 

```python
###  now that we have trained our algorithm.It's time to make some predictions.
```

```python
"print(X_test)"   #  Testing data - - In hours
"y_pred = regressor.predict(X_test)"   ##  predicting the scores
```

```python
##  comparing actual vs predicted
"df  = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})"
"print(df)"
```

```python
"score_pred = np.array([9.25])\n",
"score_pred = score_pred.reshape(-1,1)\n",
"predict = regressor.predict(score_pred)\n",
"print(\"No of hours = {}\".format(9.25))\n",
"print(\"Predicted Score ={}\".format(predict[0]))"
```

###   Evaluating the model

```python
###  the final step is to evaluate the performance of algorithm.
###  This step is particularly important to compare how well different 
###  algorithms  perform on a particular dataset.
###  from simplicity  here, we have chosen the mean square error.There are many such metrics...
```

```python
"from sklearn import metrics \n",
"print('Mean Absolute Error:', \n",
"    metrics.mean_absolute_error(y_test,y_pred))"
```
