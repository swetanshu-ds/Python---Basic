```python
import shutil
```

```python
l = []
for root, dirs, files in os.walk("F:\\"):
    for i in files:
        if i.endswith(".zip"):
            filename = os.path.join(root,i)
            dest = "F:\\zip"
            try:
                shutil.copy(filename,dest)
            except:
                l.append(filename)
            
            
```

```python
l = []
for root, dirs, files in os.walk("F:\\"):
    for i in files:
        if i.endswith(".zip"):
            filename = os.path.join(root,i)
            dest  = "F:\\newzip"
            if os.path.isdir(dest)==False:
                os.mkdir("F:\\newzip")
            
            try:
                shutil.copy(filename,dest)
            except:
                l.append(filename)
                print(filename)
            
```
