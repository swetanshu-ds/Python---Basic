# object oriented programming 

###  method 

```python
class circle:
    pi = 3.14
    def __init__(self,radius,area):
        self.radius = radius
        self.area = radius*radius*circle.pi
    def setradius(self,new_radius,new_area):
        self.radius = new_radius
        self.area = new_radius*new_radius*self.pi
    def circumference(self,radius):
        return self.radius*2*circle.pi
    
c = circle(1,1)
print("radius is",c.radius)
print("area is ",c.area)
print("area is ",c.circumference(1))
```

#  inheritance

```python
class Animal:
    def __init__(self):
         print("Animal was created")
    def whoamI(self):
          print("Animal")
    def eat(self):
          print("eating")
```

```python
class Dog(Animal):
    def __init__(self):
        Animal.__init__(self) 
        print("Dog was created")
    def whoamI(self):
        print("Dog")
    def bark(self):
        print("dog barks")
```

```python
d = Dog()
print(Animal.whoamI("huskie"))
d.bark()
d.whoamI()
d.eat()
```

##  polymorphism

```python
###  in this deifferent classes can share the same  method 
###  name and  those methods can  be called from the same
###     place , even though variety of differnt objects 
###    can be passed in it
```

```python
class Dog:
    def __init__(self,name):
        self.name = name
    def speak(self):
        return  self.name + ' barks Woof'
```

```python
class Cat:
    def __init__(self,name):
        self.name = name
    def speak(self):
        return self.name +  ' says meow '
```

```python
d = Dog("alex")
c = Cat("felix")
print(d.speak())
print(c.speak())
```
