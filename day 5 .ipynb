```python
#  creating die class
from random import randint
class Die():
    def __init__(self,num_sides = 6):
        self.num_sides = num_sides
    def roll(self):
        return  randint(1,self.num_sides)
```

```python
#  from die  import Die
die = Die()

results = []

for roll_num in range(10):
    result = die.roll()
    results.append(result)

    # analyze the result

print(results)

frequencies = []

for value in range(1,die.num_sides + 1):
    frequency = results.count(value)
    frequencies.append(frequency)

print(frequencies)    
```

#  making histogram

```python
import pygal
hist = pygal.Bar()

hist.add('D6',frequencies)
hist.render_to_file('abc.svg')
```

```python
##  steps for all machine learning models
# 1. importing the libraries
# 2. Importing the dataset
# 3.  spliting the data set into training set and test set
#  4. training model on the training set
# 5. predicting the test set result
#  6. visualizing the test set result
```

```python
# import library
import tensorflow as tf
```

```python
# import dataset
mnist = tf.keras.datasets.mnist
```

```python
# divide the dataset into training and test dataset
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
#  how data looks
print(x_train[0])
```

```python
#  visualization 
import matplotlib.pyplot as plt
plt.imshow(x_train[0],cmap= plt.cm.binary)
plt.show()
```

```python
print(y_train[17])
```

```python
#  Normalize
x_train = tf.keras.utils.normalize(x_train,axis = 1)
x_test =  tf.keras.utils.normalize(x_test,axis = 1)
```

```python
# how data looks after normalization
print(x_train[0])
```

```python
#  visualization after normalization
import matplotlib.pyplot as plt
plt.imshow(x_train[0],cmap= plt.cm.binary)
plt.show()
```

###  to create the model

```python
model = tf.keras.models.Sequential()
```

```python
##  input into 1* 784
```

```python
model.add(tf.keras.layers.Flatten())
```

```python
#  adding few layers
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))
```

```python
# output layer
model.add(tf.keras.layers.Dense(10,activation=tf.nn.softmax))


model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossetropy',metrics = ['accuracry'])
model.fit(x_train,y_train,epochs = 3)
```
