```python
def askforint():
    while True:
        try:
            a = int(input("enter an integer"))
        except Exception as e :
            print("this is my error msg",e)
            continue
        else:
            print("person has entered a correct value")
            break 
        finally:
            print("close this value")
```

```python
askforint()
```

```python
def askforint1():
    while True:
        try:
            a = int(input("enter an integer"))
        except FileNotFoundError as e :
            print("this is my error msg",e)
            continue
        except IOError as e:
            print(e)
        else:
            print("person has entered a correct value")
            break 
        finally:
            print("close this value")
```

```python
askforint1()
```

```python
def askforint2():
    while True:
        try:
            a = int(input("enter an integer"))
        except FileNotFoundError as e :
            print("this is my error msg",e)
        except IOError as e:
            print(e)
        except ValueError as e:
            print(e)
        else:
            print("person has entered a correct value")
            break 
        finally:
            print("close this value")
```

```python
askforint2()
```

```python
def askforint3():
    while True:
        try:
            a = int(input("enter an integer"))
            c = 8/a
        except FileNotFoundError as e :
            print("this is my error msg",e)
        except IOError as e:
            print(e)
        except ValueError as e:
            print(e)
        except Exception as e:
            print("this is coming  from parent class",e)
        except ArithmeticError as e:
            print("this is a arithmetic error",e)
        except ZeroDivisionError as e :
            print(e)
        else:
            print("person has entered a correct value")
            break 
        finally:
            print("close this value")
```
